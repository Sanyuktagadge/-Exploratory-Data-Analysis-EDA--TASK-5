```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Saujanya Gadge\OneDrive\Desktop\cleaned data_train.csv')
df.head()
```

```python
# Describe, Info, Value Counts

print(df.describe())


print(df.info())


print(df['SURVIVED'].value_counts())
print(df['PCLASS'].value_counts())
```

```python
# Pairplot for selected COLUMN
sns.pairplot(df[['AGE', 'FARE', 'PARCH', 'SIBSP', 'SURVIVED']], hue='SURVIVED')
plt.show()
```

```python
 # Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df[['AGE', 'FARE', 'PARCH', 'SIBSP', 'SURVIVED']].corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
#HISTOGRAM FOR AGE

df['AGE'].hist(bins=20)
plt.title('AGE Distribution')
plt.xlabel('AGE')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
#Boxplot of Fare by Survival

sns.boxplot(x='SURVIVED', y='SEX', data=df)
plt.title('Fare by Survival')
plt.show()
```

```python
#Scatterplot of Fare vs Age colored by Survival

sns.scatterplot(x='AGE', y='FARE', hue='SURVIVED', data=df)
plt.title('Fare vs Age by Survival')
plt.show()
```
